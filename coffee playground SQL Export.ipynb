### Question: How could a meaningful split of the content for a SQL-DB structure look like to fullfill SQL relation integrity?
### Task: Split the data and create sub-dataframes for the SQL-tables

## 1. Prepare the data for creation of SQL-tables and Create sub-dataframes

```python
df_coffee = df[['Unnamed: 0', 'ID', 'Country of Origin', 'Variety', 'Processing Method', 'Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 'Clean Cup', 'Sweetness', 'Overall', 'Defects', 'Total Cup Points', 'Moisture Percentage', 'Category One Defects', 'Quakers', 'Color', 'Category Two Defects']].copy()
```

```python
df_coffee.info()
```

```python
df_owner = df[['Unnamed: 0', 'Owner']].copy()
```

```python
df_owner.head
```

```python
df_farm = df[['Unnamed: 0', 'Farm Name', 'Lot Number', 'Mill', 'ICO Number', 'Altitude', 'Region', 'Producer', 'Number of Bags', 'Bag Weight', 'In-Country Partner', 'Harvest Year', 'Grading Date']].copy()
```

```python
df_certification = df[['Unnamed: 0', 'Certification Body', 'Certification Address', 'Certification Contact']].copy()
```

```python
df_country = df[['Unnamed: 0', 'Country of Origin']].copy()
```

## 2. Add PrimaryKey for SQL-DB compatiblity:

### Renaming the 'Unnamed: 0' column to match the ID columns of SQL tables


```python
df_coffee.rename(columns={'Unnamed: 0': 'CoffeeID'}, inplace=True)
```

```python
df_coffee.head(1)
```

```python
df_owner.rename(columns={'Unnamed: 0': 'OwnerID'}, inplace=True)
df_farm.rename(columns={'Unnamed: 0': 'FarmID'}, inplace=True)
df_certification.rename(columns={'Unnamed: 0': 'CertificationID'}, inplace=True)
df_country.rename(columns={'Unnamed: 0': 'CountryID'}, inplace=True)
```

```python
df_owner.head(1)
```

```python
df_farm.head(1)
```

```python
df_certification.head(1)
```

```python
df_country.head(1)
```

## 3. Export new sub-dataframes to new CSV files and SQL-DB

### A. Export each dataframe into its own CSV-file:

```python
df_coffee.to_csv('coffee.csv', index=False)
```

```python
ls -lsa coffee.csv
```

```python
df_owner.to_csv('owner.csv', index=False)
df_farm.to_csv('farm.csv', index=False)
df_certification.to_csv('certification.csv', index=False)
df_country.to_csv('country.csv', index=False)
```

```python
ls -lsa *.csv
```

### B. Export each dataframe to a table of the SQL-DB:

```python
engine = create_engine('sqlite:////Users//thedude/dev/git/material-dashboard-pro-django-main/db.sqlite3')
```

```python
df_coffee.to_sql('Coffee', engine, if_exists='replace', index=False)
```

#### Test if the SQL-DB has been successfully written:

```python
df_sql_coffee = pd.read_sql_table('Coffee', engine)
```

```python
df_sql_coffee.head()
```

```python
df_owner.to_sql('Owner', engine, if_exists='replace', index=False)
df_farm.to_sql('Farm', engine, if_exists='replace', index=False)
df_certification.to_sql('Certification', engine, if_exists='replace', index=False)
df_country.to_sql('Country', engine, if_exists='replace', index=False)
```

```python
df_sql_owner = pd.read_sql_table('Owner', engine)
df_sql_farm = pd.read_sql_table('Farm', engine)
df_sql_certification = pd.read_sql_table('Certification', engine)
df_sql_country = pd.read_sql_table('Country', engine)
```
